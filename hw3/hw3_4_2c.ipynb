```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn import tree
from sklearn import preprocessing
```

```python
PATH = 'C:/Users/Harsh/Desktop/task4_2/'
```

```python
train_data = pd.read_csv(PATH + 'train.csv')

test_data = pd.read_csv(PATH + 'test.csv')
```

```python
train= train_data.apply(preprocessing.LabelEncoder().fit_transform)
test= test_data.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
dt = DecisionTreeClassifier()
```

```python
feature_cols=['Outlook', 'Temperature', 'Humidity', 'Windy']
```

```python
x=train[feature_cols]
y=train.Label
```

```python
dt = DecisionTreeClassifier()
dt.fit(x, y)
```

```python
x_test=test[feature_cols]
```

```python
y_pred=dt.fit(x,y).predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y, y_pred))
```
